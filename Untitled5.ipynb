```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/IIoT_Smart_Parking_Management.csv')

# Display the first 5 rows of the DataFrame
print("First 5 rows of the dataset:")
display(df.head())

# Display concise summary of the DataFrame
print("\nDataFrame Information:")
df.info()

# Display descriptive statistics of the DataFrame
print("\nDescriptive Statistics:")
display(df.describe(include='all'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical columns for outlier detection
numerical_cols = df.select_dtypes(include=['number']).columns

# Plot box plots for numerical columns to visualize outliers
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(5, 5, i + 1) # Adjust subplot grid based on the number of numerical columns
    sns.boxplot(y=df[col])
    plt.title(col)
    plt.tight_layout()
plt.show()

print("\nNumerical columns in the dataset:")
print(numerical_cols)
```

```python
# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Identify numerical columns
numerical_cols = df.select_dtypes(include=['number']).columns

# Handle outliers using the IQR method (capping)
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("Outliers handled using IQR capping method for numerical columns.")

# Display descriptive statistics after outlier handling
print("\nDescriptive Statistics after Outlier Handling:")
display(df.describe(include='all'))
```
